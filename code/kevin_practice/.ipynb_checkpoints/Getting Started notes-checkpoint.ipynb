````python
import numpy as np
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit import execute
````

# Building the circuit

````python
q = QuantumRegister(3, 'q')    
circ = QuantumCircuit(q)
````

By default, each qubit in the register is initilialized to 0

Making the GHZ state:
    1. A Hadamard gate on qubit 0 (puts it in superposition)
    2. A controlled-Not (Cx) between qubit 0 and 1
    3. A controlled-Not operation between qubit 0 and 2

````python
circ.h(q[0])
circ.cx(q[0], q[1])
circ.cx(q[0], q[2])
````

## visualize the circuit

````python
circ.draw(output='mpl')
````

## Statevector_simulator 
Most common backend in Qiskit Aer. Returns the quantum state which is a complex vector of dimensions $2^{n}$ where $n$ is the number of qubits

````python
from qiskit import BasicAer
````

# Note: Qiskit uses the notation where the $Q_{n}$ qubit is on the left side of the tensor product
\begin{equation} 
Q_{n} \tensor \ellipses Q_{1} \tensor Q_{0}$
\end{equation}

````python
backend_BA = BasicAer.get_backend('statevector_simulator')
````

````python
job_BA_1 = execute(circ, backend_BA)
````

````python
job_BA_1.status()
````

````python
job_BA_1.result()
````

````python
result_BA_1 = job_BA_1.result()
````

### statevector of the quantum circuit above:

````python
output_state = result_BA_1.get_statevector(circ, decimals=3)
````

````python
from qiskit.tools.visualization import plot_state_city
plot_state_city(output_state)
````

## Unitary backend
- works only provided all teh elements in the circuit are unitary operations
- backend calculates the $2^{n} x 2^{n}$ matrix representing the gates in the quantum circuit

````python
# run the quantum circuit on a unitary simulator backend
backend_unitary = BasicAer.get_backend('unitary_simulator')
job_unitary = execute(circ, backend_unitary)
from qiskit.tools.monitor import job_monitor
job_monitor(job_unitary)
result_unitary = job_unitary.result()
````

````python
print(result_unitary.get_unitary(circ, decimals=3))
````

## OpenQASM backend
To simulate a circuit that includes measurement, we need to add measurements to the original circuit above, and use a different Aer backend:

````python
# Create a classical register with 3 bits
c = ClassicalRegister(3, 'c')
# create a quantum circuit
meas = QuantumCircuit(q, c)
meas.barrier(q)
# map the quantum measuremnet to the classical bits
meas.measure(q, c)

# The Qiskit circuit object supports 'composition' using the addition operator
qc = circ + meas

# draw the circuit
qc.draw(output='mpl')
````

Each run of this circuit will yield bitstring 000 or 111. To build statistics about the distribution of the bitstrings (estimate P(1000)), we need to repeat the circuit many times. The number of times the circuit is repeated can be specified in the ```execute``` function, via the ```shots``` keyword

### Begin: use Aer's qasm_simulator

````python
backend_qasm_sim = BasicAer.get_backend('qasm_simulator')

# execute the circuit on the qasm simulator
# set the number of times to repeat the circuit
num_shots = 1024
job_qasm = execute(qc, backend_qasm_sim, shots=num_shots)

# grab the results of the job_qasm
result_qasm = job_qasm.result()
````

````python
# access the counts 
counts_qasm_job = result_qasm.get_counts(qc)
print('Num counts qasm_job:', counts_qasm_job)
````

````python
from qiskit.tools.visualization import plot_histogram
from qiskit.tools.visualization import plot_bloch_vector
plot_histogram(counts_qasm_job)
````

````python
# find the least busy 
from qiskit.providers.ibmq import least_busy
large_enough_devices = IBMQ.backends(filters=lambda x: x.configuration().n_qubits >3 and not x.configuration().simulator)
print('large enough:', large_enough_devices)
````

````python
ideal_backend_IBMQ = least_busy(large_enough_devices)
print('ideal backend:', ideal_backend.name())
````

````python
from qiskit.tools.monitor import job_monitor
ibmq_job_shots = 1024
max_credits = 1   # maximum number of credits to spend on executions

job_exp = execute(qc, backend=ideal_backend_IBMQ, max_credits=max_credits)
job_monitor(job_exp)
````

````python
result_exp = job_exp.result()
counts_exp = result_exp.get_counts(qc)
plot_histogram(counts_exp)
````
