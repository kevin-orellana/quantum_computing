# Overview

```python
import numpy as np
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit import BasicAer, execute
from qiskit.quantum_info import Pauli, state_fidelity, basis_state, process_fidelity
```

```python
q0 = QuantumRegister(2, 'q0')
c0 = ClassicalRegister(2, 'c0')
q1 = QuantumRegister(2, 'q1')
c1 = ClassicalRegister(2, 'c1')
q_test = QuantumRegister(2, 'q0')
print(q0.name, q0.size)
```

```python
q0==q_test   # what?
```

## Creating circuits

```python
circuit = QuantumCircuit(q0, q1)
```

```python
circuit.x(q0[1])
circuit.x(q1[0])
circuit.draw(output='mpl')
```

```python
circuit_2 = QuantumCircuit()
circuit_2.add_register(q0)
circuit_2.add_register(q1)
circuit_2.x(q0[1])
circuit_2.x(q1[0])
circuit_2.draw(output='mpl')
```

```python
from copy import deepcopy
q3 = QuantumRegister(2, 'q3m')
circuit_3 = deepcopy(circuit)
circuit_3.add_register(q3)
circuit_3.draw(output='mpl')
```

## Concatenating circuits
Useful when concatenating circuits to create a new one. A use case includes when one circuit has no measurements and the final circuit represents a measurement?

```python
measurement_1 = QuantumCircuit(q0, q1, c0, c1)
measurement_1.measure(q0, c0)
measurement_1.measure(q1, c1)
qc_measurement1 = circuit + measurement_1
```
